# Домашнее задание 12

## Задача 1 ― «Менеджер расходов»

Нужно создать класс ExpenseTracker, который помогает вести учёт личных расходов:

добавлять траты,

просматривать историю,

считать расходы за определенный период или для категории,

сохранять и загружать данные из файла.
Формат файла expenses.txt:

Дата;Категория;Описание;Сумма
Примеры функций:

save_to_file() - cохранить все текущие расходы обратно в expenses.txt,

add_expense(date, category, description, amount),

get_total_by_category(category) - подсчитать, сколько всего потрачено в заданной категории,

get_total_by_day(date) - подсчитать расходы за конкретный день,

show_all() - вывести все расходы красиво построчно:

    2025-11-07 | Еда          | Обед в кафе           | 450₽
    
    2025-11-08 | Развлечения  | Кино                  | 600₽

Номер l Казино - 3 попытки

```python
import random

total_players = 1000
winners = 0

for player in range(total_players):
    attempt1 = random.randint(1, 10)
    attempt2 = random.randint(1, 10)
    attempt3 = random.randint(1, 10)
    if attempt1 == attempt2 or attempt1 == attempt3 or attempt2 == attempt3:
        winners += 1
percentage = (winners / total_players) * 100

print("Количество победителей:", winners)
print("Процент победителей:", round(percentage, 2), "%")
```

Номер ll Колесо фортуны

```python
import random
import matplotlib.pyplot as plt
sectors = ['Подарок', 'Повтори', 'Шанс', 'Ничего', 'Сюрприз']
results = {'Подарок': 0, 'Повтори': 0, 'Шанс': 0, 'Ничего': 0, 'Сюрприз': 0}
for spin in range(100):
    selected = random.choice(sectors)
    results[selected] += 1
labels = list(results.keys())
sizes = list(results.values())
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Результаты 100 вращений колеса удачи')
plt.axis('equal')
plt.show()
print("Результаты 100 вращений:")
for sector, count in results.items():
    print(f"{sector}: {count} раз(а)")
```

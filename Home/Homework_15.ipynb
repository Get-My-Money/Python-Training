# Домашнее задание 15

## Задача 1 ― «Профиль пользователя + История входов + Анализ истории входов»

Нужно создать и сохранить в profile.json профиль пользователя с помощью функции create_profile что-то типа по его выбору:

    {
    "username": "student",
    "created_at": "дата и время",
    "settings": {
        "theme": "dark",
        "notifications": true
    },
    "login_history": []
    }

Затем при каждом запуске программы функции look_profile дополнить файл профиля:

    1. Добавить в login_history текущую дату и время
    
    2. Сохранить изменения в файл
    
    3. Вывести информацию профиля

Так же используя функцию login_history, нужно посчитать:

    1. Сколько входов было сегодня
    
    2. Сколько входов было за последние 7 дней
    
    3. Вывести результат

# Домашнее задание 4

## Задача 1 ― «Четные элементы по нечётным индексам»

1. Считайте одной строкой числа через пробел в список.

2. Постройте новый список, содержащий все элементы из оригинального с нечётными индексами (1, 3, 5, …), значения которых при этом четные.

3. Выведите все элементы нового списка через пробел и их количество.

 Пример:

    Ввод: 
    
    10 20 30 41 50 60

    Вывод:

    [20, 60]

    2

## Задача 2 ― «Разность соседей»

1. На вход поступает список целых через пробел.

2. Постройте новый список длины len(orig_list)-1, где new_list[i] = orig_list[i+1] - orig_list[i]

3. Выведите новый список.

Пример:

    Ввод: 
    
    5 3 8 10

    Вывод:

    [-2, 5, 2]

## Задача 3 ― «Поворот списка»

1. Считайте список из целых через пробел и целое число k во второй строке.

2. «Поверните» список вправо на k позиций (каждый элемент сместится вправо, конец переходит в начало).

3. Выведите получившийся список.

Пример:

    Ввод:

    1 2 3 4 5

    2

    Вывод:

    [4, 5, 1, 2, 3]

## ЗАДАНИЕ 4*. Вложенные списки

1. Попросите пользователя ввести название фильма (строка).
2. Попросите пользователя ввести год выпуска (число).
3. Попросите пользователя указать собственную оценку фильма по шкале от 1 до 10.
4. Создайте список, внутри которого будет три вложенных списка, и в каждом из этих вложенных списков должна храниться пара вида: «тип/название поля» и «значение». Пример:
 [
  ["Название", "Властелин колец"],
  ["Год", 2001],
  ["Оценка", 9]
]
Заполняйте эти данные на основе ответов пользователя, а не вручную!
5. Добавьте в конец получившегося списка ещё один произвольный элемент (любого типа).
6. С помощью среза возьмите элементы с индексами 2 по 4 включительно из итогового списка и выведите их на экран.

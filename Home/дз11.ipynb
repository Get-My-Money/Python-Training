```python
#Домашнее задание 10
#Задача 1 ― «Set и Dict Comprehensions»
names = ["Анна", "Иван", "Олег", "Мария", "Слава"]
grades = [[4, 5, 5], [3, 4, 3], [5, 5, 4], [4, 4, 4], [5,5,5]]
subjects = ["математика", "физика", "информатика"]
student_data = {}

for name, grade_list in zip(names, grades):
    average_grade = sum(grade_list) / len(grade_list)
    student_data[name] = float(f"{average_grade:.1f}") 

print("Все студенты:", student_data)
student_data = {name: avg for name, avg in student_data.items() if avg > 4.0}
print("Студенты со средним баллом > 4.0:", student_data)
excellent_students = []

for name, grade_list in zip(names, grades):
    if all(grade == 5 for grade in grade_list):
        excellent_students.append(name)
student_data["отличники"] = excellent_students

print("Итоговый словарь:", student_data)

#Задача 2 - «Итератор для обратного обхода списка»
class ReverseIterator:
    def __init__(self, data):
        self.data = data
        self.index = len(data) - 1  
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.index >= 0:
            value = self.data[self.index]
            self.index -= 1
            return value
        else:
            raise StopIteration

it = ReverseIterator([10, 20, 30, 50, 5])
print(next(it))  
print(next(it))  
print(next(it))
print(next(it))
print(next(it))
```

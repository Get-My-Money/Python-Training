## Задача A ―  «Система учёта товаров в магазине»

Нужно создать класс StoreInventory, который будет управлять товарами магазина, хранящимися в текстовом файле inventory.txt.
Система должна уметь:

    читать данные из файла,
    
    хранить товары в структуре (например, словарь),
    
    добавлять, изменять и удалять товары,
    
    искать и анализировать информацию,
    
    сохранять обновлённые данные обратно в файл.

Формат файла inventory.txt, где каждая строка — отдельный товар:

    ID;Название;Категория;Цена;Количество

Примеры функций:

    save_to_file() - сохранить текущее состояние словаря self.items обратно в файл,

    add_item(name, category, price, quantity) - добавить новый товар,

    remove_item(item_id) - удалить товар по ID,

    find_by_name(keyword) - найти товары, в названии которых встречается указанное слово (регистр не учитывать),

    find_by_category(category) - вывести все товары определённой категории,

    change_quantity(item_id, new_quantity) - изменить количество товара на складе по ID.

```python
file_name = "B:/Python projects/Python-Training/Files/l_21/inventory.txt"
class StoreInventory:
    def __init__(self, file_name):
        self.file_name = file_name
        self.local_db = self.file_processing(file_name)
    def add_item(self, name, category, price, quantily):
        existed_ids = list(self.local_db.keys())
        for new_id in range(1, max(existed_ids)+2):
            if new_id not in existed_ids:
                self.local_db[new_id] = [name, category, price, quantily]
                break
                
    def remove_item(self, item_id):
        del self.local_db[item_id]

    def find_by_name(self, keyword):
        existed_ids = list(self.local_db.keys())
        for odinakovo in range(len(self.local_db)):
            if keyword in self.local_db[existed_ids[odinakovo]][0]:
                print(self.local_db[existed_ids[odinakovo]][0])
                

    
                
    def file_processing(self, file_name):
        slovarb = {}
        with open(file_name, 'r', encoding="utf8") as file:
            for line in file:
                line = line[:-1]
                a = line.split(";")
                a[3] = int(a[3])
                a[4] = int(a[4])
                slovarb[int(a[0])] = a[1:]
        return slovarb 




a = StoreInventory(file_name)
print(a.local_db.keys())
a.add_item("Макларен", "Машины", 100000000, 120)
a.remove_item(3)
a.remove_item(1)
a.remove_item(2)
a.add_item("Макларенf", "Машины", 100000000, 120)
a.add_item("Макларен", "Машины", 100000000, 120)
a.add_item("МакларенFFFFFFFFFFFFF", "Машины", 100000000, 120)
print(a.local_db)
a.find_by_name("Макларен")
```

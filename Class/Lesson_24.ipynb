## Тип datetime — дата + время

```python
# Текущие дата и время и атрибуты
dt = datetime.now()

print(dt.year)
print(dt.month)
print(dt.day)
print(dt.hour)
print(dt.minute)
print(dt.second)
```

```python
# Создание вручную
dt = datetime(2025, 3, 10, 14, 30, 0)
print(dt)
```

```python
# Форматирование
print(dt.strftime("%Y-%m-%d %H:%M:%S"))
print(dt.strftime("%d.%m.%Y %H:%M"))

# Парсинг строки
dt = datetime.strptime("2025-03-10 14:30", "%Y-%m-%d %H:%M")
print(dt)
print(type(dt))
```

```python
# Объединение date и time
new_time = datetime.combine(date(2025, 3, 10), time(14, 30))
print(new_time)
```

## timedelta — разница времени

```python
delta = timedelta(days=32, hours=5, minutes = 6, seconds = 16)
print(delta)
```

```python
# Арифметика
now = datetime.now()
later = now + timedelta(days=32, hours=5, minutes = 6, seconds = 16)
before = now - timedelta(hours=2)
print(now)
print(later)
print(before)

d1 = datetime(2025, 3, 10)
d2 = datetime(2025, 3, 1)

print((d1 - d2).days)
```

## Микрозадание 1 — время

Нужно написать программу, которая:
    
    1. Получает текущую дату и время
    
    2. Выводит их в формате: Сегодня: 10.03.2025, текущее время: 14:32.

    3. Считает и выводит:
    
        а) какая дата будет через 7 дней
    
        б) сколько часов прошло с начала текущего дня

```python
datee = date.today()
now = datetime.now()
current_time = now.time()
current_ttime = current_time.strftime("%H:%M")
print(f"Сегодня: {datee}, текущее время: {current_ttime}.")
dd = timedelta(days=7)
print(datee + dd)
d1 = current_time.hour
print(d1)
```

## Модуль json

```python
# Базовые операции: json.dumps и json.loads
data = {
    "city": "Moscow",
    "temperature": 22,
    "rain": False
}

json_string = json.dumps(data)
print(type(json_string))
print(json_string)

parsed = json.loads(json_string)
print(type(parsed))
print(parsed)
```

```python
# Работа с файлами
data = {
    "name": "Anna",
    "age": 14
}

with open("user.json", "w", ) as file:
    json.dump(data, file)
print(data)

with open("user.json", "r") as file:
    data = json.load(file)
print(data)
```

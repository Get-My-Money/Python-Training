## Задача A ― «Дедлайн-проверка»

Есть дедлайн (Введён пользователем).

Программа должна:

    1. Запросить дедлайн в формате "год-месяц-день час:минута"(например 2025-03-10 14:30).

    2. Проверить, прошёл ли дедлайн и сообщить об этом
    
    3. Если нет — показать:
    
        a) Cколько дней
        
        b) Cколько часов
        
        c) Cколько минут осталось

Пример:

    До дедлайна осталось: 1 день 3 часа 12 минут

```python
while True:
    try:    
        dead_line = datetime.strptime(input("Введите строку в виде Год-Месяц-День Час:Минуты"), "%Y-%m-%d %H:%M")
        break
    except:
        None
time_now = datetime.now()

dead_line_date = str(dead_line).split()[0] #2027-03-10 14:30
time_now_date = str(time_now).split()[0]
dt = dead_line_date.split("-")
d_now = time_now_date.split("-")

time_time_now = str(time_now).split()[1]
time_dead_line = str(dead_line).split()[1]

tm_now = time_time_now.split(":")
tl_dead = time_dead_line.split(":")

for i in range(3):
    dt[i] = int(dt[i])
    d_now[i] = int(d_now[i])
print(dt)
print(d_now)

for i in range(2):
    tm_now[i] = int(tm_now[i])
    tl_dead[i] = int(tl_dead[i])


dead_line_date_td = datetime(dt[0],dt[1],dt[2])
time_now_date_td = datetime(d_now[0],d_now[1],d_now[2])
time_dead_line_t = time(tm_now[0],tm_now[1])
time_dead_line_t = time(tl_dead[0], tl_dead[1])
print(time_now_date_td)
print(dead_line_date_td)
if time_now > dead_line:
    print("Дедлайн прошел") 
else:
    rem_days = (dead_line_date_td - time_now_date_td).days
    b = ((dead_line - time_now).total_seconds() - (dead_line_date_td - time_now_date_td).days*86400)/3600
    print(b)
    if b > 0:
        hours_rem = int(b)
        min_rem = (b - hours_rem) * 60
    else:
        b+=24
        hours_rem = int(b)
        rem_days -= 1
        min_rem = (b - hours_rem) * 60
    print(f"До ДедЛайна осталось: {rem_days} дня, {hours_rem} часов и {int(min_rem)} минуты.")
```

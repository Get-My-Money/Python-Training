## Задача B ― «Секундомер»

Нужно реализовать простой секундомер для пользователя.

Секундомер должен уметь:

    1. Запускаться
    
    2. Ставиться на паузу
    
    3. Продолжаться после паузы
    
    4. Показывать текущее прошедшее время по запросу пользователя
    
    5. Корректно считать время (с точностью до долей секунды)

    6. Реализовать сброс секундомера

```python
#0 - Время сброшенно до 0; 1 - Секундомер активен; 2 - Секундомер на паузе
counted_time = 0
condition = 0
look_time = 0
global condition,counted_time,look_time
def Start():
    global condition,counted_time,look_time
    if condition == 1:
        print("Секундомер активен")
    if condition in [0,2]:
        condition = 1
        look_time = t.time()
def Pause():
    global condition,counted_time,look_time
    if condition == 2:
        print("Секундомер на паузе")
    if condition == 0:
        print("Секундомер уже сброшен")
    if condition == 1:
        print("Секундомер активен")
        condition = 2
        counted_time = t.time() - look_time
        look_time = 0
def Reset():
    global condition,counted_time,look_time
    print("Время сброшенно до 0")
    condition = 0
    look_time = 0
    counted_time = 0
    
def Show_time():
    global condition,counted_time,look_time
    if condition == 0:
        print("Засеченное время равно нулю")
    if condition == 1:
        print(f"Время : {t.time() - look_time + counted_time}")
    if condition == 2:
        print(f"Время : {counted_time}")
```

```python
Start()
t.sleep(1)
Show_time()
Reset()
Show_time()
Start()
```

```python
Show_time()
Pause()
Show_time()
Start()
t.sleep(2)
Show_time()
Reset()
Show_time()
```

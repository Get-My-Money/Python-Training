## Задача B ― «Телепорт-квадрат»

Условие:

    1. Создать Canvas
    2. Нарисовать квадрат
    3. При клике мыши (<Button-1>):
       Квадрат перемещается в точку клика центр квадрата должен совпадать с кликом

    Доп. условия: не выходить за границы Canvas, менять цвет при каждом телепорте

```python
global color, canvas, rectangle, color_db

window = tk.Tk()
window.geometry("900x600")
window.resizable(False, False)
canvas = tk.Canvas(window, width=900, height=600)
canvas.pack()
rectangle = canvas.create_rectangle(40, 40, 80, 80, width = 3)
color = -1
color_db = ["red", "green", "blue","yellow", "purple"]

def move(event):
    global color, canvas, rectangle, color_db
    object_cords = canvas.coords(rectangle)
    object_cords_xc = (object_cords[0] + object_cords[2]) / 2
    object_cords_yc = (object_cords[1] + object_cords[3]) / 2
    move_x = event.x - object_cords_xc
    move_y = event.y - object_cords_yc
    if object_cords_xc+move_x <=23:
        move_x = 23 - object_cords_xc
    if object_cords_xc+move_x >=877: 
        move_x = 877 - object_cords_xc
    if object_cords_yc+move_y <=23:
        move_y = 23 - object_cords_yc
    if object_cords_yc+move_y >=577:
        move_y = 577 - object_cords_yc
        
    color+=1
    if color > 4:
        color = 0
        
    canvas.itemconfig(rectangle, fill=color_db[color])
    canvas.move(rectangle, move_x, move_y)
canvas.bind("<Button-1>", move)
window.mainloop()
```

## Задача А — Функции в функции: динамический фильтр + вычисления

Дан список пользователей в виде словарей:

    users = [
        {"name": "Аня", "age": 17, "city": "Москва"},
        {"name": "Марк", "age": 22, "city": "СПб"},

        {"name": "Лена", "age": 19, "city": "СПб"},

    ]

Требуется создать функцию filter_users(users, min_age=None, city=None), которая должна:

    Создать вложенную функцию check(user), которая:

        проверяет возраст, если указан min_age;
        
        проверяет город, если указан city.
        
    Вернуть:

        отфильтрованный список пользователей;

        обобщённую статистику, которую считает другая внутренняя функция build_stats(filtered).

        количество найденных пользователей,

        сколько уникальных городов среди найденных.

```python
users = [
    {"name": "Аня", "age": 17, "city": "Москва"},
    {"name": "Марк", "age": 22, "city": "СПБ"},
    {"name": "Олег", "age": 30, "city": "Казань"},
    {"name": "Таня", "age": 15, "city": "Москва"},
    {"name": "Лена", "age": 19, "city": "СПБ"},
    {"name": "Игорь", "age": 42, "city": "Казань"}
]
def filter_users(users, min_age=None, city=None):
    users_filtered = []
    def check(user):
        for user_info in users:
            user_age = user_info["age"]
            user_city = user_info["city"]
            if min_age == None or (type(user_age) == int and user_age >= 0 and user_age < 120 and type(min_age) == int and  user_age >= min_age):
                if city == None or (type(user_city) == str and user_city == city):
                    users_filtered.append(user_info)
    def build_stats(filtered):
        count_users = 0
        unique_cities = 0
        a = set()
        medium_age = 0
        for user_info in filtered:
            count_users += 1
            medium_age += user_info["age"]
            a.add(user_info["city"])
        unique_cities = len(a)
        if count_users > 0:
            medium_age = medium_age / count_users
        else:
            medium_age = None
        print("Количество найденных пользователей:", count_users)
        print("Кол-во уникальных городов:", unique_cities)
        print("Средний возраст:", medium_age)
    

        
    check(users)
    print(users_filtered)
    build_stats(users_filtered)
    return users_filtered
a = filter_users(users, 12, None)
```

## Задача B — Словарь частот слов по тексту

Пользователь задаёт текст, нужно написать функцию word_count(text), которая возвращает словарь вида {слово: количество}.

Требования:
    
    1. убрать запятые и точки;
    
    2. привести к нижнему регистру;
    
    3. разделять по пробелам.

```python
user_input = input()
def word_count(text):
    text = list(text)
    znak = [".", ",", "!", "?", ";", ":"]
    for i in range(len(text)):
        if text[i] in znak:
            text[i] = " "
    text = "".join(text)
    text = text.lower()
    text = text.split()
    slovarb = {}
    for word in text:
        if word not in slovarb:
            slovarb[word] = 1
        else:
            slovarb[word] += 1
    return slovarb
print(word_count(user_input))
```

## Задача C — Работа со списками: уникальные элементы

Создать функцию unique_sorted(lst), которая:

    1. принимает список, например [4, 2, 2, 7, 4, 1, 1, 9];
    
    2. убирает повторы;
    
    3. сортирует список;
    
    4. возвращает новый, чистый список.

    5. нельзя использовать функции sort/sorted и множества.

```python
num = [4, 2, 2, 7, 4, 1, 1, 9, 8, 3, 5, 6, 0]
def unique_sorted(lst):
    unique_num = []
    for digit in lst:
        if digit not in unique_num:
            unique_num.append(digit)
    for j in range(len(unique_num)):
        for i in range(len(unique_num) - 1 - j):
            if unique_num[i] > unique_num[i+1]:
                unique_num[i], unique_num[i+1] = unique_num[i+1], unique_num[i]
    return unique_num
print(unique_sorted(num))
```

Даны строки:

Создать функцию parse_records(data), которая превращает каждую строку в словарь вида:

    {
      "name": "...",
      "age": ... (int),
      "city": "..."
    }

И возвращает список таких словарей.

```python
data = [
    "Иванов Иван;29;Москва",
    "Петров Петр;31;Самара",
    "Сидорова Анна;24;Казань",
    "Искандеров Андрей;48;Орск",
    "Андреева Ольга;15;Таганрог",
]
```

```python
data = [
    "Иванов Иван;29;Москва",
    "Петров Петр;31;Самара",
    "Сидорова Анна;24;Казань",
    "Искандеров Андрей;48;Орск",
    "Андреева Ольга;15;Таганрог",
]
def parse_records(data):
    date_list = []
    for info in data:
        slovarb = {}
        slovarb["name"] = info.split(";")[0].split()[1]
        slovarb["age"] = info.split(";")[1]
        slovarb["city"] = info.split(";")[2]
        print(f"Имя: {slovarb["name"]}, Возраст: {slovarb["age"]}, Город: {slovarb["city"]}")
        date_list.append(slovarb)
    return date_list
a = parse_records(data)
print(a)
```

```python
from datetime import datetime, date, time, timedelta
success = {}
max_counter = 0
user_name = ""
for i in range(len(logs)-1):
    for j in range(len(logs)-1-i):         
        if datetime.strptime(logs[j]["time"], "%d/%m/%Y %H:%M:%S") > datetime.strptime(logs[j+1]["time"], "%d/%m/%Y %H:%M:%S"):
            logs[j]["time"], logs[j+1]["time"] = logs[j+1]["time"], logs[j]["time"]
            
for i in range(len(logs)):
    if logs[i]["status"] == "success":
        try:
            success[logs[i]["user"]] += 1
        except:
            success[logs[i]["user"]] = 1
        if max_counter < success[logs[i]["user"]]:
            max_counter = max(max_counter, success[logs[i]["user"]])
            user_name = logs[i]["user"]
    else:
        try:
            success[logs[i]["user"]] = 0
        except:
            success[logs[i]["user"]] = 0
```

## Задача B ― Аномальные интервалы времени

Нужно написать программу, которая:

    1. Посчитает интервалы между соседними событиями.
    
    2. Найдёт интервалы, которые больше среднего в 3 раза.
    
    3. Вернёт индексы этих «аномальных» разрывов.

```python
times = [3, 5, 7, 9, 10, 12, 40, 42, 45, 90, 91, 93, 150, 151, 152, 153, 300, 450, 451, 800, 1200, 1205, 1210, 2000]
sr = times[-1] - times[0]
el = len(times)-1
sr = sr/el
interv = []
interv_an = []
for i in range(len(times)-1):
    interv.append(times[i+1] - times[i])
    if interv[i] > sr:
        interv_an.append(i+1)
print(sr)
print(interv_an)
```

## Задача C ― Цепочки зависимостей

Нужно написать программу, которая:

    1. Определяет длину самой длинной цепочки.
    
    2. Выведет эту цепочку от начала до конца.

Пример:

    relations = [
        ("A", "B"),
        ("B", "C"),
        ("C", "D"),
        ("E", "F")
    ]
    
    Это означает A → B → C → D.

```python
relations = [
    ("Core", "Auth"),
    ("Core", "Database"),
    ("Auth", "Session"),
    ("Auth", "OAuth"),
    ("OAuth", "Google"),
    ("OAuth", "GitHub"),
    ("Database", "Postgres"),
    ("Database", "Redis"),
    ("Session", "JWT"),
    ("Session", "Cookies"),
    ("Core", "API"),
    ("API", "REST"),
    ("API", "GraphQL"),
    ("REST", "v1"),
    ("REST", "v2"),
    ("GraphQL", "Schema"),
    ("GraphQL", "Resolvers"),
    ("Resolvers", "UserResolver"),
    ("Resolvers", "ProductResolver"),
]
tree = {}
for i in range(len(relations)):
    try:
        tree[relations[i][0]] += [relations[i][1]]
    except:
        tree[relations[i][0]] = [relations[i][1]]
print(tree, end="\n")
```

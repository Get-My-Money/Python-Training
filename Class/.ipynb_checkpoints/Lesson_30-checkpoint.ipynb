## Анимационный цикл

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=500, height=400)
canvas.pack()

ball = canvas.create_oval(50, 150, 100, 200, fill="blue")


def game_loop():
    canvas.move(ball, 3, 0)
    window.after(20, game_loop)


game_loop()

window.mainloop()
```

## Скорость (vx, vy)

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=500, height=400)
canvas.pack()

ball = canvas.create_oval(50, 150, 100, 200, fill="blue")

vx = 5
vy = 0.5

def game_loop():
    canvas.move(ball, vx, vy)
    window.after(50, game_loop)

game_loop()

window.mainloop()
```

## Отскок от стен

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=500, height=400)
canvas.pack()

ball = canvas.create_oval(50, 150, 100, 200, fill="blue")

vx = 5
vy = 3

def game_loop():
    global vx, vy

    canvas.move(ball, vx, vy)

    x1, y1, x2, y2 = canvas.coords(ball)

    if x1 <= 0 or x2 >= 500:
        vx = -vx

    if y1 <= 0 or y2 >= 400:
        vy = -vy

    window.after(20, game_loop)

game_loop()

window.mainloop()
```

## Управление скоростью с клавиатуры

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=500, height=400)
canvas.pack()

ball = canvas.create_oval(50, 150, 100, 200, fill="blue")

vx = 5
vy = 3

def game_loop():
    global vx, vy

    canvas.move(ball, vx, vy)

    x1, y1, x2, y2 = canvas.coords(ball)

    if x1 <= 0 or x2 >= 500:
        vx = -vx

    if y1 <= 0 or y2 >= 400:
        vy = -vy

    window.after(60, game_loop)

def control(event):
    global vx, vy

    if event.keysym == "Up":
        vy -= 1
        print(vx, vy)

    if event.keysym == "Down":
        vy += 1
        print(vx, vy)

    if event.keysym == "Left":
        vx -= 1
        print(vx, vy)

    if event.keysym == "Right":
        vx += 1
        print(vx, vy)


window.bind("<Key>", control)

game_loop()

window.mainloop()
```

## Столкновение объектов

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=500, height=400)
canvas.pack()

ball = canvas.create_oval(50, 150, 100, 200, fill="blue")
enemy = canvas.create_rectangle(300, 200, 350, 250, fill="red")

vx = 5
vy = 3

def game_loop():
    global vx, vy

    canvas.move(ball, vx, vy)

    x1, y1, x2, y2 = canvas.coords(ball)

    if x1 <= 0 or x2 >= 500:
        vx = -vx

    if y1 <= 0 or y2 >= 400:
        vy = -vy

    if check_collision(ball, enemy):
        vx = -vx
        vy = -vy

    window.after(60, game_loop)

def check_collision(obj1, obj2):
    x1, y1, x2, y2 = canvas.coords(obj1)
    x3, y3, x4, y4 = canvas.coords(obj2)

    return x1 < x4 and x2 > x3 and y1 < y4 and y2 > y3

def control(event):
    global vx, vy
    print(vx, vy)

    if event.keysym == "Up":
        vy -= 1

    if event.keysym == "Down":
        vy += 1

    if event.keysym == "Left":
        vx -= 1

    if event.keysym == "Right":
        vx += 1


window.bind("<Key>", control)

game_loop()

window.mainloop()
```

## Canvas tags

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=500, height=400)
canvas.pack()

for i in range(3):
    canvas.create_oval(50*i, 50, 50*i+40, 90, fill="red", tags="enemy")

def game_loop():
    global vx, vy

    canvas.move("enemy", 2, 0)

    window.after(60, game_loop)

game_loop()

window.mainloop()
```

## Несколько движущихся объектов

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=500, height=300, bg="white")
canvas.pack()

balls = []

for i in range(3):
    obj = canvas.create_oval(50*i, 50, 50*i+40, 90, fill="blue")

    balls.append({
        "id": obj,
        "vx": 3 + i,
        "vy": 2
    })


def game_loop():
    for b in balls:
        canvas.move(b["id"], b["vx"], b["vy"])

    window.after(50, game_loop)

game_loop()

window.mainloop()
```

## Мини-задача: «Отбивай шарик»

Условие:

    1. Создать: шарик (круг), платформу (прямоугольник снизу)
    2. Движение шарика: шарик двигается автоматически (vx, vy) и отскакивает от левой, правой и верхней стены
    3. Платформа: управляется стрелками ← →, двигается только по X, не выходит за границы экрана
    4. Столкновение: если шарик касается платформы, то он должен отскочить вверх (меняется vy)
    5. Проигрыш: если шарик коснется нижней стены, выводим надпись "GAME OVER"

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=1000, height=600, bg="white")
canvas.pack()

ball= canvas.create_oval(50, 50, 100, 100, fill="red")
rectangle = canvas.create_rectangle(10 , 530, 150, 560, fill="green")
s = 0
vx,vy = 10,10

def game():
    global vx, vy,s
    canvas.move(ball, vx , vy)
    x1, y1, x2, y2 = canvas.coords(ball)

    if x1 <= 0 or x2 >= 1000:
        vx = -vx

    if y1 <= 0:
        vy = -vy
    if y2 > 600:
        text= canvas.create_text(470,280, text="Game over", font =("Arial",20), fill ="red")
        stext = canvas.create_text(470,220, text=f"Всего {s} очков", font =("Arial",20), fill ="green")
        return
    
    if check_collision(ball, rectangle):
        vy = -vy
        s+=1
    window.after(60, game)

def control(event):
    x1, y1, x2, y2 = canvas.coords(rectangle)

    if x1 <= 0:
        canvas.move(rectangle, 10, 0 )
    if x2>=1000:
        canvas.move(rectangle, -10, 0 )
        
    if event.keysym == "Right":
        canvas.move(rectangle, 10, 0 )
        
    if event.keysym == "Left":
        canvas.move(rectangle, -10, 0)


window.bind("<Key>", control)

def check_collision(obj1, obj2):
    x1, y1, x2, y2 = canvas.coords(obj1)
    x3, y3, x4, y4 = canvas.coords(obj2)

    return x1 < x4 and x2 > x3 and y1 < y4 and y2 > y3

window.mainloop()
```

```python
import tkinter as tk 
```

## Создание окна

```python
window = tk.Tk()

window.mainloop() # цикл обработки событий
```

## Настройка окна

```python
window = tk.Tk()

window.title("Моя первая программа")

window.geometry("900x600")

window.resizable(False, False)

window.mainloop()
```

## Label (текст)

```python
window = tk.Tk()

label = tk.Label(window, text="Привет Tkinter!")

label.pack()

window.mainloop()
```

## Кнопка (Button)

```python
window = tk.Tk()

def hello():
    print("Кнопка нажата")

button = tk.Button(window, text="Нажми меня", command=hello)

button.pack()

window.mainloop()
```

## Поле ввода

```python
window = tk.Tk()

entry = tk.Entry(window)
entry.pack()


def show_text():
    text = entry.get()
    print(text)


button = tk.Button(window, text="Показать текст", command=show_text)
button.pack()

window.mainloop()
```

## Canvas — поле для рисования

```python
window = tk.Tk()

window.geometry("900x600")

canvas = tk.Canvas(window, width=900, height=600, bg="white")
canvas.pack()

window.mainloop()
```

## Создание объектов

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=500, height=300, bg="white")
canvas.pack()

canvas.create_line(50, 50, 200, 50)

canvas.create_rectangle(50, 80, 200, 150, width = 8)

canvas.create_oval(250, 80, 350, 150, outline="blue")

canvas.create_text(150, 200, text="Hello Canvas", fill = "green")

window.mainloop()
```

## Объекты Canvas

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=400, height=300, bg="white")
canvas.pack()

ball = canvas.create_oval(100, 100, 150, 150, fill="blue")

print(ball) # ball — это ID объекта

window.mainloop()
```

## Перемещение объектов

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=400, height=300, bg = "white")
canvas.pack()

ball = canvas.create_oval(100, 100, 150, 150, fill="red")


def move_right():
    canvas.move(ball, 20, 0)

def move_left():
    canvas.move(ball, -20, 0)


button1 = tk.Button(window, text="->", command=move_right)
button2 = tk.Button(window, text="<-", command=move_left)
button1.pack()
button2.pack()

window.mainloop()
```

## Получение координат объекта

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=400, height=300)
canvas.pack()

ball = canvas.create_oval(100, 100, 150, 150, fill="blue")

print(canvas.coords(ball))

canvas.move(ball, 20, 0)

print(canvas.coords(ball))

window.mainloop()
```

## Управление с клавиатуры(bind)

```python
window = tk.Tk()

canvas = tk.Canvas(window, width=400, height=300)
canvas.pack()

ball = canvas.create_oval(100, 100, 150, 150, fill="green")


def move(event):

    if event.keysym == "d":
        canvas.move(ball, 10, 0)

    if event.keysym == "a":
        canvas.move(ball, -10, 0)

    if event.keysym == "w":
        canvas.move(ball, 0, -10)

    if event.keysym == "s":
        canvas.move(ball, 0, 10)


window.bind("<Key>", move)

window.mainloop()
```

## Базовая анимация

```python
import tkinter as tk  # обязательно импортируем модуль

window = tk.Tk()

canvas = tk.Canvas(window, width=400, height=300)
canvas.pack()

ball = canvas.create_oval(100, 100, 150, 150, fill="green")

def move_ball(step=0):
    if step < 25:          
        canvas.move(ball, 10, 0)
        window.after(500, move_ball, step + 1) 

button = tk.Button(window, text="go", command=move_ball)
button.pack()

window.mainloop()
```

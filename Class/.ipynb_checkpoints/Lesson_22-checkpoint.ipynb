## Задача A ― "Библиотека цитат”

Нужно создать программу-класс, который управляет коллекцией вдохновляющих цитат, их авторами и темами. Все данные хранятся в текстовом файле quotes.txt.
Система должна уметь:

    загружаеть цитаты,
    
    анализировать тексты,
    
    искать закономерности,
    
    строить статистику.

Формат файла quotes.txt:

    Автор;Тема;Текст цитаты

Примеры функций:

    save_to_file() - сохранить текущее состояние словаря self.items обратно в файл,

    get_longest_quote() - возвращает самую длинную цитату (по количеству символов),

    get_shortest_quote() - возвращает самую короткую цитату (по количеству символов),

    count_words() - общее количество слов во всех цитатах, среднюю длину цитаты в словах, топ-5 самых частых слов,

    authors_rating() - рейтинг авторов по “весомости”, где вес = сумма длины всех цитат,

    topic_popularity() - определяет популярность тем,

    estimate_mood() - определяет “эмоциональное настроение” автора или темы(позитивные слова: “жизнь”, “любовь”, “мечта”, негативные: “проблема”, “страх”, “ошибка”).

```python
file_name = "B:/Python projects/Python-Training/Files/l_22/quotes.txt"
class Quotes_Library:
    def __init__(self, file_name):
        self.file_name = file_name
        self.items = {}
        self.save_to_library()


    def authors_rating(self):
        slovarb = {}
        for id in self.return_ids:
            try:
                slovarb[self.items[id]["author"]] += len(self.items[id]["quote"])
            except:
                slovarb[self.items[id]["author"]] = len(self.items[id]["quote"])
        
    def save_to_file(self):
        pass

    def save_to_library(self):
        id = 0
        with open(file_name, 'r', encoding="utf8") as file:
            for line in file:
                a = line[:-1].split(";")
                self.items[id] = {"author": a[0], "theme" : a[1], "quote" : a[2]}
                id += 1
                
    def return_ids(self):
        return list(self.items.keys())


    def get_longest_quote(self):
        max_length = 0
        max_len_quote = -1
        idss = self.return_ids()
        for id in idss:
            if len(self.items[id]["quote"]) > max_length:
                max_length = len(self.items[id]["quote"])
                max_len_quote = self.items[id]["quote"]
        return max_len_quote
        
    def get_shortest_quote(self):
        min_length = 10000000
        min_len_quote = -1
        idss = self.return_ids()
        for id in idss:
            if len(self.items[id]["quote"]) < min_length:
                min_length = len(self.items[id]["quote"])
                min_len_quote = self.items[id]["quote"]
        return min_len_quote
            

    
a = Quotes_Library(file_name)
for id in a.items:
    print(id, a.items[id])
print(a.get_longest_quote())
print(a.get_shortest_quote())
```
